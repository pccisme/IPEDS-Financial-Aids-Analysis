```python
def correlation(regex):    
    df_year = df.filter(regex=regex)

    df_charges = df_year.iloc[:,0:26]
    df_aids = df_year.iloc[:, 26:]
    charges_columns = df_charges.columns
    aids_columns = df_aids.columns

    corr_index = {}

    for charge in charges_columns:
        for aid in aids_columns:
            col1 = df_charges[charge]
            col2 = df_aids[aid]
            corr = col1.corr(col2)
            if corr == corr:
                corr_index[charge + ' / ' + aid] = float(corr)

    df_corr = pd.DataFrame(corr_index, index=['value']).T
    df_corr_sorted = df_corr.sort_values(by='value', ascending=False)
    
    return df_corr_sorted
```

```python
regex = '2017|2017-18|SFA1718'
corr = correlation(regex)
corr
```

```python
regex = '2016|2016-17|SFA1617'
corr = correlation(regex)
corr
```
